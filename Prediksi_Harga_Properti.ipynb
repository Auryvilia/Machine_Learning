```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# 1. Upload dataset
print("Upload file CSV yang berisi data penjualan mobil")
uploaded = files.upload()

# 2. Load dataset
filename = list(uploaded.keys())[0]
data = pd.read_csv(filename)
```

```python
# Extract data for Year_Built and Price
x_year = data['Year_Built'].values  # Fitur independen (tahun pembangunan)
y_price = data['Price'].values  # Target (harga properti)

# Define the least squares function
def least_squares(x, y):
    n = len(x)
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_x2 = np.sum(x**2)
    sum_xy = np.sum(x * y)

    # Calculate coefficients
    a = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
    b = (sum_y - a * sum_x) / n
    return a, b

# Calculate model parameters
a, b = least_squares(x_year, y_price)
print(f"Model regresi linear: Price = {a:.2f} * Year_Built + {b:.2f}")

# Predict Price for a specific year
year_input = 2025  # Contoh tahun untuk prediksi
predicted_price = a * year_input + b
print(f"Prediksi harga properti untuk tahun {year_input}: {predicted_price:.2f}")

# Visualization: Year vs Price
plt.figure(figsize=(10, 6))

# Scatter plot for original data
plt.scatter(x_year, y_price, color='blue', label='Data asli')

# Regression line
x_line = np.linspace(min(x_year), max(x_year), 100)
y_line = a * x_line + b
plt.plot(x_line, y_line, color='red', label='Garis regresi')

# Plot prediction point
plt.scatter(year_input, predicted_price, color='green', label=f'Prediksi: {predicted_price:.2f}')

plt.xlabel('Tahun Pembangunan (Year Built)')
plt.ylabel('Harga Properti (Price)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Extract data for Year_Built and Price
x_year = data['Year_Built'].values  # Fitur independen (tahun pembangunan)
y_price = data['Price'].values  # Target (harga properti)

# Define the least squares function
def least_squares(x, y):
    n = len(x)
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_x2 = np.sum(x**2)
    sum_xy = np.sum(x * y)

    # Calculate coefficients
    a = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
    b = (sum_y - a * sum_x) / n
    return a, b

# Calculate model parameters
a, b = least_squares(x_year, y_price)
print(f"Model regresi linear: Price = {a:.2f} * Year_Built + {b:.2f}")

# Predict prices for a range of years
start_year = 2000
end_year = 2030
years = np.arange(start_year, end_year + 1)
predicted_prices = a * years + b

# Create bar chart
plt.figure(figsize=(12, 6))
plt.bar(years, predicted_prices, color='skyblue', edgecolor='blue')

plt.title('Prediksi Harga Properti Berdasarkan Tahun Pembangunan', fontsize=14)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Harga Properti (Price)', fontsize=12)
plt.xticks(years, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show predicted price for each year
for i, price in enumerate(predicted_prices):
    plt.text(years[i], price, f'{price:.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```
